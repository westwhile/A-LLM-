# 03 因子构建

输入：样例行情、估值、财务、行业和事件数据。输出：`reports/notebook_outputs/factor_coverage.csv`。

```python
from pathlib import Path
import sys
ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = next(p for p in ROOT.parents if (p / 'src').exists())
sys.path.insert(0, str(ROOT / 'src'))
from ashare_factor_research.data.data_loader import LocalDataLoader
from ashare_factor_research.pipeline import build_factor_panel
data = LocalDataLoader(ROOT / 'data' / 'sample').load_all()
factor_panel, factor_cols = build_factor_panel(data, horizon=20)
coverage = factor_panel[factor_cols].notna().mean().rename('coverage').reset_index()
coverage.columns = ['factor', 'coverage']
out_dir = ROOT / 'reports' / 'notebook_outputs'
out_dir.mkdir(parents=True, exist_ok=True)
coverage.to_csv(out_dir / 'factor_coverage.csv', index=False)
print(factor_panel.shape)
print(coverage.to_string(index=False))
```
