输入：样例 pipeline 输出。输出：`drawdown.csv`、`turnover.csv`、`industry_exposure.csv`。当前覆盖回撤、换手、成本和行业暴露聚合。

```python
from pathlib import Path
import sys
ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = next(p for p in ROOT.parents if (p / 'src').exists())
sys.path.insert(0, str(ROOT / 'src'))
from ashare_factor_research.pipeline import run_sample_pipeline
result = run_sample_pipeline(ROOT / 'data' / 'sample', ROOT / 'reports' / 'figures')
nav = result['nav'].copy()
nav['running_max'] = nav['nav'].cummax()
nav['drawdown'] = nav['nav'] / nav['running_max'] - 1.0
print(nav[['trade_date', 'nav', 'drawdown', 'turnover', 'holding_count']].tail().to_string(index=False))
print(result['industry_exposure'].tail().to_string(index=False))
```
